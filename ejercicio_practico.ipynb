```python
# 34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. 
# Los métodos disponibles son: crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol. 
# El objetivo es implementar estos métodos para manipular la estructura del árbol.

class Arbol:
    # MÉTODO CONTRUCTOR: se ejecuta al crear un objeto Arbol
    def __init__(self):
        self.tronco = 1
        self.ramas = []

    # Método 2: para hacer crecer el tronco en una unidad
    def crecer_tronco(self):
        self.tronco += 1

    # Método 3: para añadir una nueva rama de longitud 1
    def nueva_rama(self):
        self.ramas.append(1)

    # Método 4: para hacer crecer todas las ramas una unidad
    def crecer_ramas(self):
        for i in range(len(self.ramas)):
            self.ramas[i] += 1

    # Método 5: para quitar una rama en una posición concreta
    def quitar_rama(self, posicion):
        if posicion >= 0 and posicion < len(self.ramas):
            self.ramas.pop(posicion)

    # Método6: para devolver la información del árbol
    def info_arbol(self):
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitud_ramas": self.ramas}


# 1. Crear un árbol
arbol = Arbol()

# 2. Hacer crecer el tronco una unidad
arbol.crecer_tronco()

# 3. Añadir una nueva rama
arbol.nueva_rama()

# 4. Hacer crecer todas las ramas una unidad
arbol.crecer_ramas()

# 5. Añadir dos nuevas ramas
arbol.nueva_rama(), arbol.nueva_rama()

# 6. Retirar la rama situada en la posición 2
arbol.quitar_rama(2)

# 7. Obtener información sobre el árbol
print(arbol.info_arbol())
```

```python
# 36. Crea la clase UsuarioBanco ,representa a un usuario de un banco con su nombre, saldo y si tiene o no cuenta corriente. 
# Proporciona métodos para realizar operaciones como retirar dinero, transferir dinero desde otro usuario y agregar dinero al saldo.

class UsuarioBanco:
    # MÉTODO CONTRUCTOR: para inicializar el usuario
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    # Método 2: para retirar dinero del saldo
    def retirar_dinero(self, cantidad):
        if not self.cuenta_corriente:
            raise Exception("El usuario no tiene cuenta corriente")
        if cantidad > self.saldo:
            raise Exception("Saldo insuficiente")
        self.saldo -= cantidad

    # Método 3: para transferir dinero desde otro usuario al actual
    def transferir_dinero(self, otro_usuario, cantidad):
        if not self.cuenta_corriente or not otro_usuario.cuenta_corriente:
            raise Exception("Ambos usuarios deben tener cuenta corriente")
        if cantidad > otro_usuario.saldo:
            raise Exception("El usuario origen no tiene saldo suficiente")
        otro_usuario.saldo -= cantidad
        self.saldo += cantidad

    # Método 4: para agregar dinero al saldo
    def agregar_dinero(self, cantidad):
        # Sumamos la cantidad al saldo
        self.saldo += cantidad


# 1. Crear dos usuarios con cuenta corriente
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 150, True)

# 2. Agregar 20 unidades de saldo a Bob
bob.agregar_dinero(20)

# 3. Transferir 80 unidades desde Bob a Alicia
alicia.transferir_dinero(bob, 80)

# 4. Retirar 50 unidades del saldo de Alicia
alicia.retirar_dinero(50)

print("Saldo Alicia:", alicia.saldo)
print("Saldo Bob:", bob.saldo)
```

```python
# 37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras , reemplazar_palabras , eliminar_palabra. 
# Estas opciones son otras funciones que tenemos que definir primero y llamar dentro de la función procesar_texto .

def contar_palabras(texto):
    palabras = texto.split()
    contador = {}
    for palabra in palabras:
        if palabra in contador:
            contador[palabra] += 1
        else:
            contador[palabra] = 1
    return contador


def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    # Usamos replace() para sustituir la palabra original por la nueva
    return texto.replace(palabra_original, palabra_nueva)


def eliminar_palabra(texto, palabra):
    # Reemplazamos la palabra por una cadena vacía
    return texto.replace(palabra, "")


def procesar_texto(texto, opcion, *args):
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        return eliminar_palabra(texto, args[0])
    else:
        return "Opción no válida"


texto = "hola mi nombre es Beatriz"

print(procesar_texto(texto, "contar"))
print(procesar_texto(texto, "reemplazar", "Beatriz", "Bea"))
print(procesar_texto(texto, "eliminar", "hola "))
```

```python
# 40. Escribe una función que tome dos parámetros: 
# figura (una cadena que puede ser "rectangulo" , "circulo" o "triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).

import math

def calcular_area(figura, datos):
    """Calcula el área de una figura geométrica según el tipo indicado.

    Args:
        figura (str): "rectangulo", "circulo" o "triangulo"
        datos (tuple): valores necesarios para calcular el área

    Returns:
        float: área de la figura"""

    if figura == "rectangulo":
        base, altura = datos
        return round(base * altura, 2)

    elif figura == "circulo":
        radio = datos[0]
        return round(math.pi * radio ** 2, 2)

    elif figura == "triangulo":
        base, altura = datos
        return round((base * altura) / 2, 2)

    else:
        return "Figura no válida"

print(calcular_area("rectangulo", (5, 3)))
print(calcular_area("circulo", (4,)))
print(calcular_area("triangulo", (6, 2)))
```

```python
# 41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el monto final de una compra en una tienda en línea, 
# después de aplicar un descuento. 

precio = input("Introduce el precio del artículo: ")

if precio.isnumeric():
    precio = int(precio)

    cupon = input("¿Tienes un cupón de descuento? (SI/NO): ")

    if cupon == "SI":
        descuento = input("Introduce el valor del cupón: ")

        if descuento.isnumeric():
            descuento = int(descuento)

            if descuento > 0:
                precio_final = precio - descuento
                print("Precio final con descuento:", precio_final)
            else:
                print("Descuento no válido. Precio final:", precio)
        else:
            print("Valor de descuento no válido. Precio final:", precio)

    elif cupon == "NO":
        print("Precio final sin descuento:", precio)

    else:
        print("Respuesta no válida. Precio final:", precio)
else:
    print("Error: el precio debe ser un número")
```
